```python
import numpy as np
import random
import thinkstats2
import thinkplot
```

```python
random_vals = np.random.random(1000)
```

```python
pmf = thinkstats2.Pmf(random_vals, label='random values')
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='random values', ylabel='PMF')
```

```python
cdf = thinkstats2.Cdf(random_vals, label='random values')
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='random values', ylabel='CDF')
```

It is evident that the values generated by random.random are uniformly distributed. The PMF shows that each value has the same probability of appearing, while the CDF shows that 20% of the values are lower than or equal to .2, 40% are lower than or equal to .4, and so on, which is to be expected from a uniformly distributed set of values between 0 and 1.
